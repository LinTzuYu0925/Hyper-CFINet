```python
from pycocotools.coco import COCO
```

```python
coco = COCO('/ssddd/TzuYu/HyperForensics++/data/ADMM-ADAM/config0/coco_meta/train.jsonl')
ann_ids = coco.getAnnIds()
count = {'Small': 0, 'eS': 0, 'rS': 0, 'gS': 0, 'Normal':0}

for ann_id in ann_ids: 
    ann = coco.loadAnns(ann_id)[0]
    if ann['area'] < 32 ** 2:
        count['Small'] += 1
        if ann['area'] < 12 ** 2: count['eS'] += 1
        elif ann['area'] < 20 ** 2: count['rS'] += 1
        elif ann['area'] < 32 ** 2: count['gS'] += 1
    elif ann['area'] < 40 * 50:
        count['Normal'] += 1

print(count)
```
